```python
import pickle
import json
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
with open('kp.pkl','rb') as file_1:
    kp = pickle.load(file_1 )

with open('scaler.pkl','rb') as file_2:
    scaler = pickle.load(file_2)
    
with open('num_col.txt','rb') as file_5:
    num_col = pickle.load(file_5)

with open('cat_col.txt','rb') as file_6:
    cat_col = pickle.load(file_6)
```

```python
data_inf = {'senior_citizen': {2: 'No', 4: 'No', 5: 'No', 8: 'No', 13: 'No'},
 'partner': {2: 'No', 4: 'No', 5: 'No', 8: 'Yes', 13: 'No'},
 'dependents': {2: 'No', 4: 'No', 5: 'No', 8: 'No', 13: 'No'},
 'tenure': {2: 2, 4: 2, 5: 8, 8: 28, 13: 49},
 'internet_service': {2: 'DSL',
  4: 'Fiber optic',
  5: 'Fiber optic',
  8: 'Fiber optic',
  13: 'Fiber optic'},
 'online_security': {2: 'Yes', 4: 'No', 5: 'No', 8: 'No', 13: 'No'},
 'online_backup': {2: 'Yes', 4: 'No', 5: 'No', 8: 'No', 13: 'Yes'},
 'device_protection': {2: 'No', 4: 'No', 5: 'Yes', 8: 'Yes', 13: 'Yes'},
 'tech_support': {2: 'No', 4: 'No', 5: 'No', 8: 'Yes', 13: 'No'},
 'streaming_tv': {2: 'No', 4: 'No', 5: 'Yes', 8: 'Yes', 13: 'Yes'},
 'streaming_movies': {2: 'No', 4: 'No', 5: 'Yes', 8: 'Yes', 13: 'Yes'},
 'contract': {2: 'Month-to-month',
  4: 'Month-to-month',
  5: 'Month-to-month',
  8: 'Month-to-month',
  13: 'Month-to-month'},
 'paperless_billing': {2: 'Yes', 4: 'Yes', 5: 'Yes', 8: 'Yes', 13: 'Yes'},
 'payment_method': {2: 'Mailed check',
  4: 'Electronic check',
  5: 'Electronic check',
  8: 'Electronic check',
  13: 'Bank transfer (automatic)'},
 'monthly_charges': {2: 53.85, 4: 70.7, 5: 99.65, 8: 104.8, 13: 103.7},
 'churn': {2: 'Yes', 4: 'Yes', 5: 'Yes', 8: 'Yes', 13: 'Yes'}}
```

```python
data_inf = pd.DataFrame(data_inf)
data_inf
```

```python
data_inf_num = data_inf[num_col]
data_inf_cat = data_inf[cat_col]
```

```python
num_scaled = scaler.transform(data_inf_num)
```

```python
data_inf_final = np.concatenate([num_scaled, data_inf_cat], axis=1)
data_inf_final = pd.DataFrame(data_inf_final, columns=['tenure', 'monthly_charges'] + cat_col)
data_inf_final = data_inf_final.infer_objects()
data_inf_final
```

```python
index_cat_columns = [data_inf_final.columns.get_loc(col) for col in cat_col]

print('Categorical columns        : ', cat_col)
print('Categorical columns index  : ', index_cat_columns)
```

```python
pred_inf = kp.predict(data_inf_final, categorical=index_cat_columns)
pred_inf
```

```python
data_inf['cluster'] = pred_inf
data_inf
```
